### Operators in Python
Python provides various operators to perform computations and logical operations. This notebook will cover different types of operators and their usage with examples.

### Boolean Values
A boolean data type represents two values: True or False. Booleans are commonly used in conditional expressions and comparisons.

```python
print(True)
print(False)
```

### Assignment Operators
Assignment operators are used to assign values to variables. For example, = assigns a value, while += adds and assigns.

```python
# Assignment Operators Examples
x = 5 # Assigning value
x += 3 # Equivalent to x = x + 3
print('x:', x)
```

### Arithmetic Operators
Arithmetic operators are used for mathematical operations like addition, subtraction, multiplication, and division.

```python
# Arithmetic Operations
a, b = 10, 3
print('Addition:', a + b)
print('Subtraction:', a - b)
print('Multiplication:', a * b)
print('Division:', a / b)
print('Modulus:', a % b)
print('Floor Division:', a // b)
print('Exponentiation:', a ** b)
```

### Comparison Operators
Comparison operators compare values and return True or False.

```python
# Comparison Operators
x, y = 5, 10
print('x > y:', x > y)
print('x < y:', x < y)
print('x == y:', x == y)
print('x != y:', x != y)
print('x >= y:', x >= y)
print('x <= y:', x <= y)
```

### Logical Operators
Logical operators (and, or, not) are used to combine multiple conditions.

```python
# Logical Operators
a, b = True, False
print('a and b:', a and b)
print('a or b:', a or b)
print('not a:', not a)
```
